# PA-GDF

Providing `BaseSchema` class to inherit from in order to validate your spatial data. 

```python
import geopandas as gpd
import shapely
from pa_gdf import BaseSchema

class MySchema(BaseSchema):

    @classmethod
    def _geometry_types(cls):
        return {shapely.Point, shapely.Polygon}
```

```python
gdf = gpd.GeoDataFrame(geometry=[
    shapely.Point(0,0),
    shapely.Point(1,1)
], crs=3857)
MySchema(gdf)
```

```python
gdf = gpd.GeoDataFrame(geometry=[
    shapely.Point(0,0),
    shapely.LineString([shapely.Point(0,0), shapely.Point(1,1)])
], crs=3857)
MySchema(gdf)
```
